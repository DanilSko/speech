This notebook provides code for parsing csv file with rules and replacing the " symbol with '«' or '»'

```python
import csv
from pandas import DataFrame
import re
```

read csv file:

```python
def read_csv(path, sep):
    return DataFrame.from_csv(path, index_col=None, sep=sep)
```

```python
df = read_csv('kavychki_6.csv', sep=";")
```

```python
df
```

read text to edit

```python
with open("testdoc.txt", 'r', encoding='utf-8') as f:
    text = f.read()
```

replace other simple types of quotes

```python
QUOTES_OPEN = '«'
QUOTES_CLOSE = '»'
SPECIFIC_QUOTES_OPEN = '„“‘„'
SPECIFIC_QUOTES_CLOSE = '“”’”'
```

```python
def replace_specific_quotes(text):
    text = text.replace(SPECIFIC_QUOTES_OPEN, QUOTES_OPEN)
    text = text.replace(SPECIFIC_QUOTES_CLOSE, QUOTES_CLOSE)
    return text
```

```python
text = replace_specific_quotes(text)
```

Function for replacing quotes

```python
def replace_quotes(rule, result_quote, text):
    try:
        text = re.sub(rule, result_quote, text)
        print(rule, "="*3, re.findall(result_quote, text))
    except Exception as e:
        print(e)
        print(rule, "="*3, "ERROR IN REG EXP")
    return text
```

Go through the cycle and replace quotes in text

```python
for ind in range(len(list(df.values))):
    rule = df.values[ind][0]
    quote = df.values[ind][1]
    text = replace_quotes(rule, quote, text)
```
